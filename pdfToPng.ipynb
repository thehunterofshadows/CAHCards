```python
from PIL import Image
output_dir = r"/home/justin/github/cahcards/pdf/cuts"
x=0
marginX=75
marginY=75
width=598
height=598
midPageBreak=165

for png_file in os.listdir(pdf_dir):
    x=x+1
    print(png_file)
    if png_file.endswith(".png"):
        i = Image.open(png_file)
        #left side of the page
        marginX=75
        for g in range (0,3):
            for z in range (0,4):
                frame2 = i.crop((marginX+(width*g),
                                 marginY+(height*z),
                                 marginX+(width*(g+1)),
                                 marginY+(height*(z+1))
                                ))
                file_loc = output_dir + "/" +"cah_og_page_"+str(x)+"_side_a_"+str(g)+str(z)+ ".png"
                frame2.save(file_loc)
                print(file_loc)
        
        #right side of the page
        marginX=2625
        for g in range (0,3):
            for z in range (0,4):
                frame2 = i.crop((marginX+(width*g),
                                 marginY+(height*z),
                                 marginX+(width*(g+1)),
                                 marginY+(height*(z+1))
                                ))
                file_loc = output_dir + "/" +"cah_og_page_"+str(x)+"_side_b_"+str(g)+str(z)+"b" + ".png"
                frame2.save(file_loc)
                print(file_loc)
```
